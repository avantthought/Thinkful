```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = 'unicode_escape')
pd.set_option('display.max_row', 200)
df
```

```python
#Remove leading and trailing whitespace

df['Publisher'] = df['Publisher'].str.strip()
df['Journal title'] = df['Journal title'].str.strip()
```

```python
#Replace double+ whitespace with single whitespace

df['Publisher'] = df['Publisher'].replace('\s+', ' ', regex=True)
df['Journal title'] = df['Journal title'].replace('\s+', ' ', regex=True)
```

```python
#remove punctuation, parenthesis, inc, ltd
#replace ampersand with and

df['Publisher'] = df['Publisher'].replace({'\.',',','/','-','Ltd','LTD','Inc','INC','\(','\)'}, ' ', regex=True)
df['Publisher'] = df['Publisher'].replace('&', ' and ', regex=True)

df['Journal title'] = df['Journal title'].replace({'\.',',','/','-','Ltd','LTD','Inc','INC','\(','\)'}, ' ', regex=True)
df['Journal title'] = df['Journal title'].replace('&', ' and ', regex=True)


#remove starting and ending whitespace
#replace doulbe+ whitespace with single space

df['Publisher'] = df['Publisher'].str.strip()
df['Journal title'] = df['Journal title'].str.strip()

df['Publisher'] = df['Publisher'].replace('\s+', ' ', regex=True)
df['Journal title'] = df['Journal title'].replace('\s+', ' ', regex=True)
```

```python
#lowercase

df['Publisher'] = df['Publisher'].str.lower()
df['Publisher'] = df['Publisher'].replace({'publications','publishing','publication','publisher',
                                          'publishers','press'}, ' ', regex=True)
df['Journal title'] = df['Journal title'].str.lower()

df['Publisher'] = df['Publisher'].str.strip()
df['Journal title'] = df['Journal title'].str.strip()

df['Publisher'] = df['Publisher'].replace('\s+', ' ', regex=True)
df['Journal title'] = df['Journal title'].replace('\s+', ' ', regex=True)
```
