```python
import numpy as np
import pandas as pd
import pandas_profiling
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('vgsales.csv')
df
```

```python
df1 = df[df.Global_Sales >= 1]
df1 = df1.drop(df1.index[[0,1]])
df1
```

```python
sns.pairplot(df1, vars=['Year',"NA_Sales","EU_Sales","JP_Sales","Global_Sales"], hue='Platform')
```

```python
sns.pairplot(df1, vars=['Year',"NA_Sales","EU_Sales","JP_Sales","Global_Sales"], hue='Genre')
```

```python
df1.describe()
```

```python
pandas_profiling.ProfileReport(df1)
```

```python
df1['NA_Ratio']=df1['NA_Sales'] / df1['Global_Sales']
df1['JP_Ratio']=df1['JP_Sales'] / df1['Global_Sales']
df1['EU_Ratio']=df1['EU_Sales'] / df1['Global_Sales']
df1
```

```python
df_wii = df1.loc[lambda df: df1['Platform'] == 'Wii',: ]
df_wii
```

```python
df_xbox360 = df1.loc[lambda df: df1['Platform'] == 'X360',: ]
df_xbox360
```

```python
df_ps3 = df1.loc[lambda df: df1['Platform'] == 'PS3',: ]
df_ps3
```

```python
plt.figure(figsize=(12,10))

plt.subplot(1,3,1)
plt.scatter(
    x=df_ps3['Year'],
    y=df_ps3['NA_Ratio'],
    color='purple',
    marker='x', s=3
)
plt.ylabel('North American Sales to Global Sales Ratio')
plt.title('PS3')

plt.subplot(1,3,2)
plt.scatter(
    x=df_wii['Year'],
    y=df_wii['NA_Ratio'],
    color='green',
    marker='x', s=3
)
plt.ylabel('North American Sales to Global Sales Ratio')
plt.title('Wii')

plt.subplot(1,3,3)
plt.scatter(
    x=df_xbox360['Year'],
    y=df_xbox360['NA_Ratio'],
    color='red',
    marker='x', s=3
)
plt.ylabel('North American Sales to Global Sales Ratio')
plt.title('XBox 360')

plt.tight_layout()
plt.show()
```
