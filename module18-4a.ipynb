```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sqlalchemy import create_engine

# Display preferences.
%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format
pd.set_option('display.max_columns', 100)

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from weatherinszeged',con=engine)

engine.dispose()

df.head(20)
```

```python
df['tempdifference'] = df['apparenttemperature'] - df['temperature']

# Y is the target variable
Y = df['tempdifference']
# X is the feature set
X = df[['humidity','windspeed']]

X = sm.add_constant(X)

# We fit an OLS model using statsmodels
results = sm.OLS(Y, X).fit()

# We print the summary results
print(results.summary())
```

The bias is 2.4381 degrees celsius, and the coefficients are -3.03 and -0.12 for humidity and windspeed respectively. All of these coefficients are statistically signficant with p-values zero. For every increase of one unit in humidity, we expect a decrease in the gap between apparent and actual temp by 3 degrees. For every increase in one unit of windspeed, we expect a decrease in the gap between apparent and actual temp by 0.12 degress.

```python
df['wind_humidity'] =  df['humidity']* df['windspeed']

# Y is the target variable
Y = df['tempdifference']
# X is the feature set
X = df[['humidity','windspeed','wind_humidity']]

X = sm.add_constant(X)

# We fit an OLS model using statsmodels
results = sm.OLS(Y, X).fit()

# We print the summary results
print(results.summary())
```

Every feature is still statistically signficant with zero p-values. The signs of the coefficients for humidity and windspeed have changed. wind_humidity has a larger coefficient.
